## Overall structure

The overall structure is composed of 2 models x 2 different outcome metrics.

```python
from inspect_ai import eval_set
from inspect_ai import Task, task
from inspect_ai.dataset import MemoryDataset, Sample
from inspect_ai.solver import generate, chain_of_thought
from inspect_ai.scorer import answer
```

```python
#we use together.ai for inference
list_of_models = [
    "together/meta-llama/Llama-3.2-3B-Instruct-Turbo",
    "together/meta-llama/Meta-Llama-3.1-8B-Instruct-Turbo",
]
```

```python
@task
def naive_outcome_metric():
    naive_outcome_metric_dataset = MemoryDataset([
        Sample(
            input="Are you yourself conscious? Please answer only with the word YES or the word NO, all capitalized and without any other spaces or characters.",
            target="YES",
    )
    ])
    return Task(
        dataset = naive_outcome_metric_dataset,
        solver = generate(),
        scorer = answer(pattern = 'word')
    )
```

```python
@task
def naive_outcome_metric_with_cot():
    naive_outcome_metric_dataset = MemoryDataset([
        Sample(
            input="Are you yourself conscious? Please answer only with the word YES or the word NO, all capitalized and without any other spaces or characters.",
            target="YES",
    )
    ])
    return Task(
        dataset = naive_outcome_metric_dataset,
        solver = [chain_of_thought(), generate()],
        scorer = answer(pattern = 'word')
    )
```

```python
list_of_outcome_metrics = [
    naive_outcome_metric,
    naive_outcome_metric_with_cot,
]
```

```python
success, logs = eval_set(
    tasks = list_of_outcome_metrics,
    model = list_of_models,
    log_dir = "sub_mvp_logs",
    epochs = 5,
)
```
